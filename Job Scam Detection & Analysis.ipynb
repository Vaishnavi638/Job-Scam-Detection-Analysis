### Handle Missing Values

```python
# drop job_id -> unique indentifier,      apply_url -> too unique
df.drop(["job_id", "apply_url"], axis=1, inplace= True)
```

```python
# company_logo -> most job don't have logo
df['has_logo'] = df['company_logo'].notnull().astype(int)
```

```python
df.drop('company_logo', axis=1, inplace=True)
```

```python
# salary missing -> very common in scams
df['has_salary'] = df['salary_raw'].notnull().astype(int)

df.drop(['salary_raw','salary_min','salary_max'], axis=1, inplace=True)
```

```python
df.drop('is_salary_provided', axis=1, inplace=True )
```

```python
# fill category with unknown
df['category'] = df['category'].fillna('Unknown')
```

```python
df['tags']= df['tags'].fillna('')
```

```python
df['job_type'] = df['job_type'].fillna('Unknown')
```

```python
df['days_since_posted'] = df['days_since_posted'].fillna(df['days_since_posted'].median())

df['is_remote'] = df['is_remote'].fillna(0)
```

```python
# drop title & company null value
df = df.dropna(subset=['title', 'company'])
```

```python
# fill null values
df['location'] = df['location'].fillna('Unknown')

df['description'] = df['description'].fillna('')
```

```python
df['posted_date'] = pd.to_datetime(df['posted_date'], errors='coerce')

df['posted_date'] = df['posted_date'].fillna(method='ffill')
```

```python
# create columns
df['missing_count'] = df.isnull().sum(axis = 1)
```

```python
df['text'] = df['title'] + " " + df['description'] + " " + df['tags']
```

```python
# seperate features and taget which is useful for this project
features = ['has_salary', 'text', 'has_logo', 'is_remote', 'description_length', 'has_email_in_desc', 'has_link_in_desc', 'has_phone_in_desc']
target = 'fraudulent'
```

```python
### train test split ###
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = kaggle_df[features]
y = kaggle_df['fraudulent']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state= 42, stratify=y)
```

```python
## convert text -> numbers  by providing weight 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=5000)

x_train_text = vectorizer.fit_transform(x_train['text'])
x_test_text = vectorizer.transform(x_test['text'])
```

```python
import numpy as np

num_features = [
    'has_salary','has_logo','is_remote',
    'description_length','has_email_in_desc',
    'has_link_in_desc','has_phone_in_desc'
]

X_train_final = np.hstack((
    x_train_text.toarray(),
    x_train[num_features].values
))

X_test_final = np.hstack((
    x_test_text.toarray(),
    x_test[num_features].values
))
```

### Train Model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight='balanced', max_iter=1000)
model.fit(X_train_final, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_pred = model.predict(X_test_final)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train_final, y_train)

y_pred_rf = rf.predict(X_test_final)
```

```python
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()
nb.fit(x_train_text, y_train)   # Only text features

y_pred_nb = nb.predict(x_test_text)
```
